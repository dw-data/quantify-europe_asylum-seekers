# Third country nationals refused entry at the external borders - annual data (rounded) [migr_eirfs]

```python
df=pd.read_csv("migr_eirfs_1_Data_mit2012.csv", encoding="latin-1")
```

```python
columns_list = df.columns

for column in columns_list:
    if column == "Value":
        pass
    else:
        print(column)
        print(df[column].value_counts())
```

```python
df.info()
```

```python
def float_conversion(strvalue):
    try:
        value=float(strvalue.replace(",", ""))
        return value
    except:
        pass
```

```python
df["parsed_value"]=df["Value"].apply(float_conversion)
```

```python
dftotal=df[(df["REASON"]=="Total")&(df["CITIZEN"]=="Total")&(df["INDIC_MG"]=="Persons refused entry")]
```

```python
dftotal.sort_values(by="parsed_value", ascending=False)
```

```python
dftotal2012 = dftotal[dftotal["TIME"]==2012]
```

```python
dftotal2012=dftotal2012[["GEO", "parsed_value"]]
```

```python
dftotal2012.columns=(["GEO", "refused2012"])
```

```python
dftotal2012
```

```python
years_list = [2013,2014,2015,2016]

dfrefused = dftotal2012

for year in years_list:
    dfrefused_temp = dftotal[dftotal["TIME"]==year]
    dfrefused_temp = dfrefused_temp[["GEO", "parsed_value"]]
    dfrefused_temp.columns = (["GEO", "refused"+str(year)])
    dfrefused = dfrefused.merge(dfrefused_temp, left_on="GEO", right_on="GEO")
```

```python
dfrefused
```

```python
dfrefused = dfrefused.drop([0])
```

```python
dfrefused
```

```python
dfrefused["aggregated_refused_entry_12-16"]=dfrefused["refused2012"]+dfrefused["refused2013"]+dfrefused["refused2014"]+dfrefused["refused2015"]+dfrefused["refused2016"]
```

```python
dfrefused
```

```python
countries_list = []
for country in dftotal["GEO"]:
    countries_list.append(country)
    
countries_set = set(countries_list)
```

```python
df=dftotal.copy()
fig, axes = plt.subplots(nrows=9, ncols=4, sharex=True, sharey=True, figsize=(15,30))
axes_list = [item for sublist in axes for item in sublist]

countries = countries_set

# subset_df = df[(df['GEO'].isin(countries))&(df["All_Decisions"]>1000)]
subset_df = df[df['GEO'].isin(countries)] 

for countryname, selection in subset_df.groupby('GEO'):
    ax = axes_list.pop(0)
    selection.plot(x='TIME', y='parsed_value', label=countryname, ax=ax, legend=False)
    ax.set_title(countryname)
    ax.set_ylim(0,200000)
    ax.set_xticklabels(["2012", " ","'13"," ","'14", " ", "'15", " ", "'16"])
    ax.tick_params(
        which='minor',
        bottom='off',
        left='off',
        right='off',
        top='off'
    )
    ax.tick_params(
        which='major',
        bottom='on',
        left='off',
        right='off',
        top='off'
    )
    #ax.grid(linewidth=0.25)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_xlabel("")

for ax in axes_list:
    ax.remove()

plt.subplots_adjust(hspace=0.55)

# plt.savefig("energy-multiples-prices-domestic.pdf", transparent = True)
```
