```python
df=df2merge.copy()
fig, axes = plt.subplots(nrows=9, ncols=4, sharex=True, sharey=True, figsize=(15,30))
axes_list = [item for sublist in axes for item in sublist]

countries = countries_set

# Filter for a minimum number of decision?
# subset_df = df[(df['GEO'].isin(countries))&(df["All_Decisions"]>1000)]
subset_df = df[df['GEO'].isin(countries)] 

for countryname, selection in subset_df.groupby('GEO'):
    ax = axes_list.pop(0)
    selection.plot(x='TIME', y='share_positive', label=countryname, ax=ax, legend=False)
    ax.set_title(countryname)
    ax.set_ylim(0,100)
    ax.tick_params(
        which='minor',
        bottom='off',
        left='off',
        right='off',
        top='off'
    )
    ax.tick_params(
        which='major',
        bottom='on',
        left='off',
        right='off',
        top='off'
    )
    #ax.grid(linewidth=0.25)
    ax.spines['left'].set_visible(False)
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)
    ax.set_xlabel("")

for ax in axes_list:
    ax.remove()

plt.subplots_adjust(hspace=0.55)

# plt.savefig("energy-multiples-prices-domestic.pdf", transparent = True)
```
